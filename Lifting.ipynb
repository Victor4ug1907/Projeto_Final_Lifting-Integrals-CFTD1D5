## Integral de Renormalização $\mathcal{J}_{3,k}$ na CFT D1-D5

A Teoria de Campos Conformes (CFT) associada ao sistema D1-D5, especificamente no ponto de Orbifold Simétrico $S_N(T^4)$, desempenha um papel central na correspondência AdS/CFT e na descrição microscópica de buracos negros em teoria de cordas. No entanto, o ponto de orbifold representa uma teoria livre. Para descrever a física gravitacional (como a dinâmica de buracos negros), é necessário mover a teoria desse ponto livre através de uma deformação marginal por um operador de deformação ${O}^{\text{int}}$. Um dos desafios centrais no estudo dessas deformações é o cálculo das dimensões anômalas (lifting) adquiridas por operadores que são protegidos (BPS) no ponto livre, mas que geralmente deixam de ser na introdução de uma deformação. O cálculo analítico dessas correções envolve a avaliação de integrais complexas do tipo Dotsenko-Fateev, cujos coeficientes crescem combinatorialmente com o índice de twist $n$ e o modo de R-simetria $k$. Este projeto foca no estudo computacional do lifting do operador quiral de twist $\sigma_n^{-}$ sob a ação de modos descendentes fracionários de simetria-R, $J_{-k/n}^3$. Se $\Delta$ é a dimensão conforme do operador quiral $\sigma_n^{-}$, a dimensão renormalizada é obtida em uma segunda ordem perturbativa por:

$$\Delta_\lambda = \Delta - \lambda^2 \frac{\pi}{2} \mathcal{J}_{3,k}(n)$$

Onde $\mathcal{J}_{3,k}(n)$ é a integral da função de correlação envolvendo os descendentes $J_{-k/n}^3\sigma^{-}_n$, definida por:

$$\mathcal{J}_{3,k}(n)=\int d^2u G(u,\bar u)$$

em que
$$ G(u)=\Big\langle J_{k/n}^3 \sigma^{-\dagger}_{(n)}(\infty)   O^{(\text{int})\dagger}_{(2)}(1) O^{(\text{int})}_{(2)}(u)J_{-k/n}^3 \sigma^-_{(n)} (0)\Big\rangle$$

A princípio devemos obter uma expressão para a função de correlação acima, e então calcular a integral do modulo quadrado da mesma. Em um trabalho recente, mostramos que essa função de correlação é obtida por:

$$G_{n}^{3;k,\pm}(y)=-\frac{1}{2}\left(\frac{b_0}{b_{\infty}}\right)^{-\frac{k}{n}}\left[k+\sum_{r=1}^{k}\frac{\mathcal{Q}_r}{y^r}+\sum_{s=1}^{k}\frac{\mathcal{R}_s}{(y-\omega_n)^s} \right]\hat G^{\{\text{prim}\}}_n(y) \tag{1}$$

em que $b_0,b_{\infty}$ são funções de $y$ e 

$$\hat G^{\{\text{prim}\}}_n(x)=\Big\langle  \sigma^{-\dagger}_{(n)}(\infty)  O^{(\text{int})\dagger}_{(2)}(t_1)  O^{(\text{int})}_{(2)}(x) \sigma^-_{(n)} (0)\Big\rangle$$


Ou seja, toda a informação dos operadores de simetria é separada da função de correlação envolvendo apenas os primários e o operador de deformação, do qual temos uma expressão conhecida para a função de corelação. 



Após uma série de manipulações, conseguimos obter a integral de correção da dimensão conforme como uma composição de integrais de dotsenko-fateev e certos coeficientes:

$$\begin{align}
	\frac{1}{\kappa}\mathcal{J}_{3,k}(n)&= \left\{\left[k^2+\frac{2k\mathcal{Q}_1}{\alpha}\right]+\sum_{j=1}^{k}\Omega_jV_j\right\}\mathcal{I}(\alpha,\beta,\gamma,\omega_n) + \sum_{j=1}^{k}\Omega_jT_j\frac{\partial}{\partial\omega_n}\mathcal{I}(\alpha,\beta,\gamma_,\omega_n) \nonumber\\
	&+\sum_{s=1}^k\left\{\frac12\Theta_{0,s}+\sum_{j=1}^{k-s}\Theta_{j,s}V_j^s\right\}\mathcal{I}(\alpha,\beta,\gamma-s,\omega_n) + \sum_{s=1}^{k}\sum_{j=1}^{k-s}\Theta_{j,s}T_j^s\frac{\partial}{\partial\omega_n}\mathcal{I}(\alpha,\beta,\gamma-s,\omega_n)\nonumber\\
	&+	\sum_{r=1}^{k}\left\{\Lambda_r+\sum_{j=1}^{k-r}\sum_{i=1}^{j}\mathcal{L}_{r,j,i}V_i^r\right\}\mathcal{I}(\alpha-r,\beta,\gamma_,\omega_n) + \sum_{r=1}^{k}\sum_{j=1}^{k-r}\sum_{i=1}^{j}\mathcal{L}_{r,j,i}T_i^r\frac{\partial}{\partial{\omega_n}}\mathcal{I}(\alpha-r,\beta,\gamma,\omega_n)
\end{align} \tag{2}$$

As funçoes $\mathcal{I}$ são conhecidas como integrais de Dotsenko-Fateev (DF), cujos parametros são definidos como
$$\alpha=1-\frac kn, \quad \beta=-\frac32, \quad \gamma=\frac kn$$
As expressões para as integrais e coeficientes serão definidas mais adiante.

É evidente que a complexidade do resultado cresce significativamente com $k$, tornando-se impraticavel de se obter a solução manualmente. Neste trabalho iremos implementar os coeficientes e integrais desta solução, a fim de se obter o comportamento da mesma e realizar as análises dos resultados obtidos. Nas proximas partes desse projeto iremos definir os coeficientes, implementá-los e realizar verificações a partir de resultados analíticos.

## Parte 1: Coeficientes Fundamentais

Para obter a solução da integral, é necesário definir uma coleção de coeficientes que atuam como um "peso" para cada integral de Dotsenko-Fateev ($\mathcal{I}$). Esses coeficientes por si só também são compostos por outros coeficientes, sendo eles:

1. $\mathcal{C}_{l,m}$:
   Esse coeficiente surge de uma simplificação de polinomios de Bell e está relacionado aos números de Lah. Resumidamente, podemos obter o coeficiente a partir de uma relação binomial entre os índices.
   $$\mathcal{C}_{\ell,m}=\frac{1}{m!}\binom{\ell - 1}{m-1}$$

2. $\mathcal{D}_{r,p,q}(k,n)$:
    O coeficiente $\mathcal{D}$ depende diretamente dos parâmetros $k$ e $n$, sendo composto pelo coeficiente $\mathcal{C}_{l,m}$ e fatoriais crescentes.
   $$\mathcal{D}_{r,p,q}(k,n)\equiv \mathcal{C}_{r,p}\mathcal{C}_{r,q}\left(\frac{k}{n}\right)_{p}\left(\frac{-k}{n}\right)_{q}$$

Nota: A notação $(x)_p$ indica o fatorial crescente, em que
    $$(x)_{p}=x(x+1)...(x+p-1)$$
    
Os dois coeficientes seguintes são a base da nossa solução geral, onde todos os outros coeficientes derivam dos mesmos. Eles são obtidos a partir de uma expansão de frações parciais nas funções de correlação. 

3. $\mathcal{Q}_r(k,n)$:
   Este coeficiente está associado à expansão a partir dos termos $y^{-r}$ de acordo com a expressão (1), e pode ser obtido por:
$$\mathcal{Q}_r(k,n) = 2q_n^2\sum_{p=r}^{k}\sum_{q=1}^{k}\mathcal{D}_{k,p,q}\zeta_r^{p,q}+\sum_{\ell=1}^{k-r}\sum_{p=r}^{k-\ell}\sum_{q=1}^{k-\ell}\mathcal{D}_{k-\ell,p,q}\ell \zeta_r^{p,q}$$

4. $\mathcal{R}_s(k,n)$:
   Este coeficiente está associado à expansão a partir dos termos $(y-\omega_n)^{-s}$ de acordo com a expressão (1), e pode ser obtido por:
$$ \mathcal{R}_s(k,n) = 2q_n^2\sum_{p=1}^{k}\sum_{q=s}^{k}\mathcal{D}_{k,p,q}\upsilon_s^{p,q}+\sum_{\ell=1}^{k-s}\sum_{p=1}^{k-\ell}\sum_{q=s}^{k-\ell}\mathcal{D}_{k-\ell,p,q}\ell \upsilon_s^{p,q}$$

Nota: Definimos $q_n$ como a carga do campo quiral em que o modo de simetria R $J^3$ está a atuar. Para o campo em que trabalharemos nesse algoritmo temos que
$$q_n=\frac{n-1}{2}$$

Por fim, os coeficientes $\zeta_i$ e $\upsilon_j$ são dados por:
$$\zeta_i^{p,q} = (-1)^{p+q}\omega_n^{i}{p+q-i-1 \choose q-1} , \quad \upsilon_j^{p,q} = (-1)^{p+q}(-\omega_n)^j{p+q-j-1 \choose p-1  }$$

em que $\omega_n=4n/(n+1)^2$

```python
import numpy as np
import scipy.special as sc
from functools import lru_cache
```

```python
##Utilizamos do pacote lru_cache que guarda valores ja calculados da função para agilidade no processamento.

## Aqui definimos o falling factorial que será útil em diversas partes do algoritmo. 
@lru_cache(maxsize=None)
def falling_factorial(a, n):
    if n == 0: return 1.0
    n_int = int(round(n))
    if n_int < 0: return 0.0
    return np.prod([(a - i) for i in range(n_int)])

## A função binomial
@lru_cache(maxsize=None)
def binomial(n, k):
    return sc.binom(n, k)

## A função que retorna o valor da carga ao quadrado, para utilizamos na definição dos coeficientes Q e R.
def qn_sq(n):
    return ((n - 1) / 2)**2

## Esse coeficiente é em certos momentos tratado como variável de derivação, e posteriormente devemos associar um valor real ao mesmo. Definimos como:
def omega(n):
    return 4 * n / (n + 1)**2
```

```python
## O coeficiente C_{l,m} 
@lru_cache(maxsize=None)
def C(ell, m):
    """
    Calcula o coeficiente C_{ell, m}.
    Fórmula: (1/m!) * binomial(ell - 1, m - 1)
    """
    if m < 0: return 0.0
    fact_m = sc.factorial(m)
    return (1.0 / fact_m) * binomial(ell - 1, m - 1)

## O coeficiente D_{r,p,q}:
@lru_cache(maxsize=None)
def D_coeff(n, k, r, p, q):
    """Calcula o coeficiente D_{r,p,q}.
    Fórmula: C(r,p)C(r,q) * (k/n)_p * (-k/n)_q
    Nota: Podemos relacionar o fatorial crescente (x)_p com o fatorial descrescente (-x)^{(p)} por
        (a)_p = (-1)^p (-a)^{(p)}
    """
    c_kp = C(r, p)
    c_kq = C(r, q)
    ff_1 = falling_factorial(-k / n, p)
    ff_2 = falling_factorial(k / n, q)
    return c_kp * c_kq *(-1)**(p+q)* ff_1 * ff_2

## As funções zeta e upsilon:
@lru_cache(maxsize=None)
def zeta(n, r, p, q):
    """Calcula zeta_r^{p,q}.
    Fórmula: (-1)^{p+q}\omega_n^r {p+q-r-1 \choose q-1  }
    """
    w_n = omega(n)
    term_omega = w_n**r
    sign = (-1)**(p + q)
    binom_term = binomial(p + q - r - 1, q - 1)
    return sign * term_omega * binom_term

@lru_cache(maxsize=None)
def upsilon(n, s, p, q):
    """Calcula upsilon_s^{p,q}.
    Fórmula: (-1)^{p+q}(-\omega_n)^s {p+q-s-1 \choose p-1  }
    """
    w_n = omega(n)
    term_omega = (-w_n)**s 
    sign = (-1)**(p + q)
    binom_term = binomial(p + q - s - 1, p - 1)
    return sign * term_omega * binom_term


## Por fim os coeficientes Q_r e R_s: 
def Q_r(n, k, r):
    q_sq = qn_sq(n)
    
    # Soma 1
    sum1 = 0.0
    for p in range(r, k + 1):
        for q in range(1, k + 1):
            sum1 += D_coeff(n,k, k, p, q) * zeta(n, r, p, q)
    term1 = 2 * q_sq * sum1
    
    # Soma 2
    sum2 = 0.0
    for ell in range(1, k - r + 1):
        inner_sum = 0.0
        t = k - ell
        for p in range(r, t + 1):
            for q in range(1, t + 1):
                inner_sum += D_coeff(n,k, t, p, q) * zeta(n, r, p, q)
        sum2 += ell * inner_sum
        
    return term1 + sum2

def R_s(n, k, s):
    q_sq = qn_sq(n)
    
    # Soma 1
    sum1 = 0.0
    for p in range(1, k + 1):
        for q in range(s, k + 1):
            sum1 += D_coeff(n, k,k, p, q) * upsilon(n, s, p, q)
    term1 = 2 * q_sq * sum1
    
    # Soma 2
    sum2 = 0.0
    for ell in range(1, k - s + 1):
        inner_sum = 0.0
        t = k - ell
        for p in range(1, t + 1):
            for q in range(s, t + 1):
                inner_sum += D_coeff(n, k,t, p, q) * upsilon(n, s, p, q)
        sum2 += ell * inner_sum
        
    return term1 + sum2
```

## 1.1 Verificação 
É de extrema importância que verifiquemos se as funções obtidas numericamente correspondem ao obtido de forma analitica. A partir da definição dos coeficientes $Q_r$ e $R_s$ obtemos para k=1 e k=2 as funções:
   1. $k=1$
        $$\mathcal{Q}_1(1,n)=-\frac{2}{n}\frac{(n-1)^2}{(n+1)^4}$$
        $$\mathcal{R}_1(1,n)=\frac{2}{n}\frac{(n-1)^2}{(n+1)^4}$$

   2. $k=2$
        $$\mathcal{Q}_1(2,n)=-\frac{4(n-1)^2}{n^3(n+1)^2}[n^2-4]-\frac{16}{n(n+1)^2}$$
        $$\mathcal{Q}_2(2,n)=\frac{8(n-1)^2}{n^2(n+1)^4}[n^2+4n+4]$$

       $$\mathcal{R}_1(2,n)=\frac{4(n-1)^2}{n^3(n+1)^2}[n^2-4]+\frac{16}{n(n+1)^2}$$
        $$\mathcal{R}_2(2,n)=\frac{8(n-1)^2}{n^2(n+1)^4}[n^2-4n+4]$$

Dessa forma, podemos implementar uma função que verifica para uma coleção de valores de $n$ se o resultado numérico é equivalente ao analítico.

```python
## Aqui implementamos os resultados analíticos
def analytical_Q1_k1(n):
    # Q_1 para k=1
    term = ((n - 1)**2) / ((n + 1)**2)
    return -(2 / n) * term

def analytical_R1_k1(n):
    # R_1 para k=1 
    return -analytical_Q1_k1(n)

def analytical_Q1_k2(n):
    # Q_1 para k=2
    # Fórmula: -(4/n^3) * [ (n-1)^2 / (n+1)^4 ] * (n^2-4) - 16/ n(n+1)^2
    pre_factor = -4 / (n**3)
    frac_term = ((n - 1)**2) / ((n + 1)**2)
    poly_term = ( (n**2) - 4)
    return pre_factor * frac_term * poly_term-16/(n*(n+1)**2)

def analytical_Q2_k2(n):
    # Q_2 para k=2
    # Fórmula: (8/n^2) * [ (n-1)^2 / (n+1)^4 ] * (n^2 + 4n +4)
    pre_factor = 8 / (n**2)
    frac_term = ((n - 1)**2) / ((n + 1)**4)
    poly_term = (4 + (n**2) + 4*n)
    return pre_factor * frac_term * poly_term

def analytical_R1_k2(n):
    # R_1 para k=2
    # Fórmula: (4/n^3) * [ (n-1)^2 / (n+1)^4 ] * (n^2-4) + 16/ n(n+1)^2
    pre_factor = 4 / (n**3)
    frac_term = ((n - 1)**2) / ((n + 1)**2)
    poly_term = ( (n**2) - 4)
    return pre_factor * frac_term * poly_term + 16/(n*(n+1)**2)

def analytical_R2_k2(n):
    # R_2 para k=2
    # Fórmula: (8/n^2) * [ (n-1)^2 / (n+1)^4 ] * (n^2-4n+4)
    pre_factor = 8 / (n**2)
    frac_term = ((n - 1)**2) / ((n + 1)**4)
    poly_term = (4 + (n**2) - 4*n)
    return pre_factor * frac_term * poly_term


def run_validation_tests():
    # Lista de valores de n para testar 
    test_n_values = [2, 3, 4, 5, 10, 100] 
    
    print(f"{'TESTE':<20} | {'n':<5} | {'Numérico':<15} | {'Analítico':<15} | {'Status'}")
    print("-" * 75)

    all_passed = True

    for n in test_n_values:
        #  Para k=1 
        # Teste Q_1 (k=1)
        val_num = Q_r(n, 1, 1) 
        val_ana = analytical_Q1_k1(n)
        # Definimos uma tolerância para um erro numérico de 1e-9
        if not np.isclose(val_num, val_ana, atol=1e-9):
            print(f"Q_1 (k=1) FAIL    | {n:<5} | {val_num:.5e}   | {val_ana:.5e}   | ERRO")
            all_passed = False
        
        # Teste R_1 (k=1)
        val_num = R_s(n, 1, 1) 
        val_ana = analytical_R1_k1(n)
        if not np.isclose(val_num, val_ana, atol=1e-9):
            print(f"R_1 (k=1) FAIL    | {n:<5} | {val_num:.5e}   | {val_ana:.5e}   | ERRO")
            all_passed = False

        # Para k=2 
        # Teste Q_1 (k=2)
        val_num = Q_r(n, 2, 1) 
        val_ana = analytical_Q1_k2(n)
        if not np.isclose(val_num, val_ana, atol=1e-9):
            print(f"Q_1 (k=2) FAIL    | {n:<5} | {val_num:.5e}   | {val_ana:.5e}   | ERRO")
            all_passed = False

        # Teste Q_2 (k=2)
        val_num = Q_r(n, 2, 2)
        val_ana = analytical_Q2_k2(n)
        if not np.isclose(val_num, val_ana, atol=1e-9):
            print(f"Q_2 (k=2) FAIL    | {n:<5} | {val_num:.5e}   | {val_ana:.5e}   | ERRO")
            all_passed = False

        # Teste R_1 (k=2)
        val_num = R_s(n, 2, 1) 
        val_ana = analytical_R1_k2(n)
        if not np.isclose(val_num, val_ana, atol=1e-9):
            print(f"Q_1 (k=2) FAIL    | {n:<5} | {val_num:.5e}   | {val_ana:.5e}   | ERRO")
            all_passed = False

         # Teste R_2 (k=2)
        val_num = R_s(n, 2, 2) 
        val_ana = analytical_R2_k2(n)
        if not np.isclose(val_num, val_ana, atol=1e-9):
            print(f"Q_1 (k=2) FAIL    | {n:<5} | {val_num:.5e}   | {val_ana:.5e}   | ERRO")
            all_passed = False

    if all_passed:
        print("-" * 75)
        print("Todos os testes passaram para os valores de n verificados.")
    else:
        print("-" * 75)
        print("Alguns valores são distintos do analítico.")
        # Em caso de erro, o proprio teste já printa o valor em que há distinção de valores


if __name__ == "__main__":
    run_validation_tests()
```

## Parte 2: Coeficientes Finais
Obtidos os coeficientes fundamentais, podemos então definir os coeficientes que compõem o resultado final da integral de renormalização da dimensão conforme. Os coeficientes são definidos da seguinte forma:

1. $\Omega_m$:

   
    Este coeficiente é o peso que define a relevância da integral de DF 'base', ou seja, sem nenhum shift nos parametros $\alpha$ ou $\gamma$. Obtemos esse coeficiente por
    $$\Omega_m\equiv2(-1)^m\left[\frac{k\mathcal{R}_m}{(\gamma)^{(m)}}+\frac{k\mathcal{Q}_m(\omega_n-1)^m}{(\alpha)^{(m)}}+\frac{k\mathcal{Q}_{m+1}(m+1)(\omega_n-1)^m}{(\alpha)^{(m+1)}}+\frac{\mathcal{Q}_{1}\mathcal{R}_{m}}{\alpha(\gamma)^{(m)}} \right]$$
    em que $(x)^{(m)}$ indica o fatorial descrescente,
    $$(x)^{(m)}=x(x-1)\cdots (x-m+1)$$



3. $\Theta_{j,s}$

   
    Este coeficiente está associado à integral de DF com um shift em $\gamma$, e pode ser definido como

   $$\Theta_{j,s}\equiv\left[\frac{2\mathcal{R}_s\mathcal{R}_{j+s}(-1)^{j}}{(\gamma-s)^{(j)}}  +  \mathbb{C}_{j,s}\right]$$

onde $\mathbb{C}_{j,s}$ é dado por
$$	\begin{align}
\mathbb{C}_{j,s}&=2(\omega_n-1)^{j+s}\mathcal{R}_s(-1)^{j}(\gamma)^{(s)}\left[\frac{\mathcal{Q}_{j+s}}{(\alpha)^{(j+s)}}+(j+s+1)\frac{\mathcal{Q}_{j+s+1}}{(\alpha)^{(j+s+1)}}\right]\nonumber\\
	&+2(\omega_n-1)^s\mathcal{R}_{j+s}\frac{(-1)^{j}(\gamma)^{(s)}(\gamma)^{(s)}}{(\gamma)^{(j+s)}}\left[\frac{\mathcal{Q}_s}{(\alpha)^{(s)}}+(s+1)\frac{\mathcal{Q}_{s+1}}{(\alpha_{\pm})^{(s+1)}}\right]
    \end{align}$$

3. $\Lambda_r$ e $\mathcal{L}_{r,j,i}$
   
   Por fim, para as integrais de DF com um shift em $\alpha$ temos dois coeficientes, definidos por
   $$\begin{align}
	\Lambda_r &\equiv \left[ |\mathcal{Q}_r|^2 + 2\sum_{j=1 }^{k-r}\frac{\mathcal{Q}_r\mathcal{Q}_{j+r}}{(\alpha-r)^{(j)}}(1-2r)^{(j)} \right]\\
	\mathcal{L}_{r,j,i} &\equiv 2 \frac{\mathcal{Q}_r\mathcal{Q}_{j+r}}{(\alpha-r)^{(j)}}{j\choose i}(-1)^i(1-2r)^{(j-i)}(\omega_n-1)^i
\end{align}$$

Precisamos portanto, definir tais coeficientes fazendo uso de seus componentes $\mathcal{Q}$ e $\mathcal{R}$ já implementados anteriormente.

```python
## Definimos aqui os parametros alpha e gamma necessários para a computação dos coeficientes
def alpha_gamma(n,k):
    """Retorna a tupla (alpha, gamma) dependente de n para cada valor de k
    alpha = 1-k/n
    gamma = k/n
    """
    return 1.0 - k/n, k/n

# Coeficiente Omega_m 
@lru_cache(maxsize=None)
def Omega_m(n, k, m):
    """
    Calcula o coeficiente Omega_m.

    """
    alpha, gamma = alpha_gamma(n,k)
    w_n = omega(n) # usa a função omega(n) já definida anteriormente
    
    # Funções Q e R necessárias:
    # Note que precisamos de Q_{m+1}, então o índice r é m+1
    val_Rm = R_s(n, k, m)
    val_Qm = Q_r(n, k, m)
    val_Qm_plus_1 = Q_r(n, k, m + 1)
    val_Q1 = Q_r(n, k, 1)
    
    # Termo comum de potência: (omega_n - 1)^m
    term_w = (w_n - 1)**m
    
    # Aqui vamos definir os denominadores que aparecem no coeficiente.
    # É importante notar que certos denominadores podem divergir, como quando k/n é um inteiro.
    # A princípio vamos supor que n>k.
    denom_gamma_m = falling_factorial(gamma, m)
    denom_alpha_m = falling_factorial(alpha, m)
    denom_alpha_m_plus1 = falling_factorial(alpha, m+1)
    
    # Verificação de segurança 
    if any(d == 0 for d in [denom_gamma_m, denom_alpha_m, denom_alpha_m_plus1]):
        print(f"Divisão por zero detectada em Omega_m (n={n}, m={m}, k={k})")
        return np.nan

    # Dentro do colchetes temos 4 termos, que são:
    
    # 1. k * R_m / (gamma)^m
    term1 = (k * val_Rm) / denom_gamma_m
    
    # 2. k * Q_m * (w_n - 1)^m / (alpha)^m
    term2 = (k * val_Qm * term_w) / denom_alpha_m
    
    # 3. k * Q_{m+1} * (m+1) * (w_n - 1)^m / (alpha)^(m+1)
    term3 = (k * val_Qm_plus_1 * (m + 1) * term_w) / denom_alpha_m_plus1
    
    # 4. Q_1 * R_m / (alpha * (gamma)^m)
    term4 = (val_Q1 * val_Rm) / (alpha * denom_gamma_m)
    
    # O resultado final é a soma dos quatro termos multiplicado por 2(-1)^m
    return 2 *(-1)**(m)* (term1 + term2 + term3 + term4)
```

```python
## Antes de definir o coeficiente \Theta precisamos definir o coeficiente \mathbb{C}. Podemos ver que este coeficiente possui duas partes 
## semelhantes, de forma que vamos definir a subfunção dos colchetes por simplificação. 

@lru_cache(maxsize=None)
def bracket_term(n, k, m, alpha_val):
    """
    Calcula o termo comum:
    [ Q_m / (alpha)^m  +  (m+1) * Q_{m+1} / (alpha)^{m+1} ]
    """
    # Coeficientes Q
    val_Qm = Q_r(n, k, m)
    val_Qm_plus_1 = Q_r(n, k, m + 1)
    
    # Denominadores
    denom_1 = falling_factorial(alpha_val, m)
    denom_2 = falling_factorial(alpha_val, m + 1)
    
    if denom_1 == 0 or denom_2 == 0:
        print(f" Divisão por zero detectada em C_j,s (n={n}, k={k})") 
        
    term1 = val_Qm / denom_1
    term2 = (m + 1) * val_Qm_plus_1 / denom_2
    
    return term1 + term2

# Função Principal C_{j,s} 
@lru_cache(maxsize=None)
def C_js(n, k, j, s):
    alpha, gamma = alpha_gamma(n,k)
    w_n = omega(n)
    term_w_minus_1 = w_n - 1
    
    # Esse coeficiente é composto por dois termos:
    # 1. 2 * (w-1)^(j+s) * R_s *(-1)^j* (gamma)^s * Bracket(j+s)
    
    ff_gamma_s = falling_factorial(gamma, s) 
    
    val_Rs = R_s(n, k, s)
    bracket_1 = bracket_term(n, k, j + s, alpha)
    
    part1 = 2 *(-1)**(j)* (term_w_minus_1**(j + s)) * val_Rs * ff_gamma_s * bracket_1
    
    # 2. 2 * (w-1)^s * R_{j+s} *(-1)^j* [(gamma)^s * (gamma)^s / (gamma)^{j+s}] * Bracket(s)
    
    val_R_js = R_s(n, k, j + s)
    ff_gamma_s = falling_factorial(gamma, s)
    ff_gamma_js = falling_factorial(gamma, j + s)
    
    gamma_fraction = (ff_gamma_s * ff_gamma_s) / ff_gamma_js
        
    bracket_2 = bracket_term(n, k, s, alpha)
    
    part2 = 2 * (term_w_minus_1**s) * val_R_js *(-1)**(j)* gamma_fraction * bracket_2
    
    return part1 + part2
```

Portanto, podemos definir o coeficiente $\Theta_{j,s}$ a partir de $\mathbb{C}$

```python
@lru_cache(maxsize=None)
def Theta_js(n, k, j, s):
    """
    Calcula Theta_{j,s}.
    """
    alpha, gamma = alpha_gamma(n,k)
    
    val_Rs = R_s(n, k, s)
    val_R_js = R_s(n, k, j + s)
    val_Cjs = C_js(n, k, j, s)
    
    denom = falling_factorial(gamma-s, j)
    
    if denom == 0:
        print(f"Divisão por zero em Theta (n={n}, j={j}, s={s}, k={k})")
        return np.nan
        
    term1 = (2 * val_Rs * val_R_js*(-1)**(j)) / denom
    
    return term1 + val_Cjs
```

Por fim, definimos os coeficientes $\Lambda_r$ e $\mathcal{L}_{r,j,i}$

```python
@lru_cache(maxsize=None)
def Lambda_r(n, k, r):
    """
    Calcula Lambda_r.
    """
    alpha, _ = alpha_gamma(n,k)
    
    val_Qr = Q_r(n, k, r)
    
    # Primeiro termo: |Q_r|^2
    term1 = val_Qr**2
    
    # Segundo termo: Soma
    sum_val = 0.0
    
    # Soma de j=1 até k-r
    for j in range(1, k - r + 1):
        
        val_Q_jr = Q_r(n, k, j + r)
        
        # Denominador: (alpha-r)^j
        denom = falling_factorial( alpha-r, j)
        
        if denom == 0:
            print(f"Divisão por zero em Theta (n={n}, j={j}, r={r}, k={k})")
            continue
            
        # Numerador: Q_r * Q_{j+r} * (1-2r)^j
        ff_2r = falling_factorial(1-2*r , j)
        
        term = (val_Qr * val_Q_jr * ff_2r) / denom
        
        sum_val += term
        
    return term1 + (2 * sum_val)
```

```python
@lru_cache(maxsize=None)
def L_rji(n, k, r, j, i):
    """
    Calcula o coeficiente L_{r,j,i}.
    """
    alpha, _ = alpha_gamma(n,k)
    w_n = omega(n)
    
    val_Qr = Q_r(n, k, r)
    val_Q_jr = Q_r(n, k, j + r)
    
    # Denominador: (alpha-r)^j
    denom = falling_factorial( alpha-r, j)
    
    if denom == 0:
        print(f"Divisão por zero em L_rji (n={n}, r={r}, j={j}, k={k})")
        return np.nan
        
    # Termos do numerador
    # Binomial(j, i)
    binom_val = binomial(j, i)
    
    # Falling factorial (1-2r)^{j-i}
    ff_2r = falling_factorial(1-2*r, j - i)
    
    # Potência (omega_n - 1)^i
    term_w = (w_n - 1)**i
    
    numerator = 2 * val_Qr * val_Q_jr * binom_val *(-1)**(i)* ff_2r * term_w
    
    return numerator / denom
```

## Parte 3: Redução de ordem 

A solução geral da integral da correção da dimensão conforme pode ser obtida em um primeiro momento como uma composição de derivadas de ordem $j$ com $j \in [1,k]$. Porém, a partir da expressão para a solução da integral de DF podemos observar que computar derivadas de ordens arbitrárias possui uma alta complexidade. Felizmente, podemos obter uma equação diferencial que relaciona uma derivada de segunda ordem da integral de DF ($\mathcal{I}(a,b,c,z)$) com derivadas de ordem 1 e 0 [1]:

$$\frac{\partial^2}{\partial{z^2}}\mathcal{I}=P(z)\frac{\partial}{\partial{z}}\mathcal{I}+Q(z)\mathcal{I}$$
em que
$$
\begin{align}
	P(z)&=\frac{\left[(a+c)-(a+b+2c)z \right]}{z(1-z)}\\
	Q(z)&=\frac{c(a+b+c+1)}{z(1-z)}
\end{align}
$$

Dessa forma, podemos expressar qualquer derivada de ordem $j$ como uma combinação de derivadas de ordem 0 e ordem 1. Ou seja,
$$\frac{\partial^j}{\partial{z^j}}\mathcal{I}=V_j(z)\mathcal{I}+T_j(z)\frac{\partial}{\partial{z}}\mathcal{I}$$

em que
$$V_{j+1}=V_j^{(1)}+T_jQ, \quad T_{j+1}=V_j+T_j^{(1)}+T_jP$$
com 
$$V_0=1,\quad V_1=0, \quad T_0=0, \quad T_1=1$$

Nota 1: A notação $(1)$ está indicando a derivada de primeira ordem aqui.

Essas expressões para $V$ e $T$ definem uma regra de recorrência que definem as ditas funções, que são então utilizadas na expressão final.

[1]: DOI: 10.1134/S1064562416060223

Precisamos portanto implementar um algoritmo que compute as funções $V$ e $T$ dados os parametros $\alpha,\beta$ e $\gamma$ além da ordem da derivada em questão. Iremos fazer isso a partir de uma abordagem semi simbólica. Primeiramente tratamos $\omega_n$ como uma variável, realizamos as devidas derivações e por fim fazemos $\omega_n=$ omega(n)

Para este passo fazemos uma simplificação importante. Veja que as funções P e Q são funções racionais, de forma que podemos escrever a derivada de ordem $j$ das mesmas como um numerador apropriado e um denominador com potencia conhecida. Ou seja, escrevemos
$$V_j = \frac{v_j}{D^j}\quad \text{e} \quad T_j = \frac{t_j}{D^j}$$

em que $D=\omega_n(\omega_n-1)$. As funções $v_j$ e $t_j$ podem ser obtidas a partir de uma simples recorrencia por
$$ \begin{align}
v_{j+1} &= (v^{(1)}D  -j v_j D^{(1)})+t_j NQ\\
t_{j+1} &= v_jD + (t^{(1)}D  -j t_j D^{(1)})+t_j NP
\end{align}$$

Onde $NQ=\gamma(a+b+c+1)$ e $NP=\left[(a+c)-(a+b+2c)\omega_n \right]$.

```python
import sympy as sym

class VT_Shifted_Generator:

    def __init__(self, n, k, r=0, s=0, beta=-1.5):
        self.w = sym.Symbol('w')
        
        alpha_base = 1.0 - k/n
        gamma_base = k/n
     
        #Note que na solução geral temos que obter as funções V e T tanto para o caso base, em que a,b,c=\alpha,\beta,\gamma,
        # quando para o caso em que há um shift nos parametros \alpha e \gamma por r e s, respectivamente
        self.alpha_eff = alpha_base - r
        self.gamma_eff = gamma_base - s
        self.beta = beta
        
        # Denominador base D = w(1-w) e sua derivada D' = 1 - 2w
        self.D_poly = sym.Poly(self.w * (1 - self.w), self.w)
        self.D_prime_poly = sym.Poly(1 - 2*self.w, self.w)
        
        # Numerador de P: [ (a+c) - (a+b+2c)w ]
        num_P_expr = (self.alpha_eff + self.gamma_eff) - \
                     (self.alpha_eff + self.beta + 2*self.gamma_eff) * self.w
        self.NP_poly = sym.Poly(num_P_expr, self.w)
        
        # Numerador de Q: [ c(a+b+c+1) ]
        num_Q_expr = self.gamma_eff * (self.alpha_eff + self.beta + self.gamma_eff + 1)
        self.NQ_poly = sym.Poly(num_Q_expr, self.w)
        
        # V_0 = 1, T_0 = 0
        self.v_polys = [sym.Poly(1.0, self.w)]
        self.t_polys = [sym.Poly(0.0, self.w)]

    def generate_up_to(self, max_j):
        # Essa definição de qual o valor maximo a ser calculado é uma maneira de reduzir o tempo de computação.
        # Sempre que essa classe for chamada ela verifica quais já foram calculados e somente calcula o que estiver faltando.
        current_max = len(self.v_polys) - 1
        if max_j <= current_max: return

        for j in range(current_max, max_j):
            v = self.v_polys[j]
            t = self.t_polys[j]
            
            v_prime = v.diff()
            t_prime = t.diff()
            
            # Recorrência dos Numeradores
            # v_{j+1} = v' D - j v D' + t NQ
            next_v = v_prime * self.D_poly - \
                     v * self.D_prime_poly * j + \
                     t * self.NQ_poly
            
            # t_{j+1} = v D + t' D - j t D' + t NP
            next_t = v * self.D_poly + \
                     t_prime * self.D_poly - \
                     t * self.D_prime_poly * j + \
                     t * self.NP_poly
            
            self.v_polys.append(next_v)
            self.t_polys.append(next_t)

    def get_numerical_values(self, w_val):
        denom_base = w_val * (1.0 - w_val)
        if abs(denom_base) < 1e-15: denom_base = 1e-15
            
        V_vals = []
        T_vals = []
        
        for j, (vp, tp) in enumerate(zip(self.v_polys, self.t_polys)):
            denom = denom_base**j
            val_v = float(vp.eval(w_val))
            val_t = float(tp.eval(w_val))
            
            V_vals.append(val_v / denom)
            T_vals.append(val_t / denom)
            
        return V_vals, T_vals
```

Novamente, é interessante realizarmos uma verificação analítica desses algoritmos. Com as definições de $V_j$ e $T_j$ pelas regras de recorrência podemos obter como exemplos de verificação, para o caso base, as funções:

$$V_0=1, T_0=0$$
$$V_1=0, T_1=1$$
$$V_2=\frac{k/2n}{\omega_n(1-\omega_n)}, \quad T_2=\frac{1+(\frac12-\frac kn)\omega_n}{\omega_n(1-\omega_n)}$$
$$V_3=\frac{\frac{5k}{4n}-\frac{k^2}{2n^2}}{\omega_n(1-\omega_n)^2}, \quad T_3=\frac{\omega_n(\frac{k^2}{n^2}+\frac34 -\frac{5k}{2n})+3(1-\frac{k}{2n})}{\omega_n(1-\omega_n)^2} $$

e com isso, definimos a função abaixo para verificar no caso de $k=1$ e $j=0,1,2,3$ se o algoritmo obtem o resultado esperado.

```python
def analytical_VT_k1(n, w):
    """Retorna listas [V0..V3] e [T0..T3] analíticas para k=1."""
    k = 1.0 
    

    V0 = 1.0
    V1 = 0.0
    V2 = (1.0/(2*n)) / (w*(1-w))
    V3 = (5.0/(4*n) - 1.0/(2*n**2)) / (w * (1-w)**2)
    
    T0 = 0.0
    T1 = 1.0
    T2 = (1.0 + (0.5 - k/n)*w) / (w*(1-w))
    
    term_livre = 3 * (1.0 - 1.0/(2*n))
    term_w = (1.0/(n**2) + 0.75 - 2.5/n) 
    T3 = (term_livre + term_w * w) / (w * (1-w)**2)
    
    return [V0, V1, V2, V3], [T0, T1, T2, T3]

def run_VT_validation_tests():

    test_n_values = [2, 3, 4, 5, 10, 20, 100]
    
    
    print(f"{'TESTE V/T (k=1)':<20} | {'n':<5} | {'Numérico':<15} | {'Analítico':<15} | {'Status'}")
    print("-" * 75)
    
    all_passed = True
    
    for n in test_n_values:
        w_val = omega(n)
        # Obtem valores numéricos via Algoritmo
        # k=1, r=0, s=0
        gen = VT_Shifted_Generator(n=n, k=1, r=0, s=0, beta=-1.5)
        gen.generate_up_to(3) # Gera até ordem 3
        V_alg, T_alg = gen.get_numerical_values(w_val)
        
        # Obtem valores analíticos
        V_ana, T_ana = analytical_VT_k1(n, w_val)
        
        # Comparação das funções V 
        for j in range(4):
            val_num = V_alg[j]
            val_ref = V_ana[j]
            
            if not np.isclose(val_num, val_ref, atol=1e-9):
                print(f"V_{j} FAIL            | {n:<5} | {val_num:.5e}    | {val_ref:.5e}    | ERRO")
                all_passed = False
        
        # Comparação das funções T 
        for j in range(4):
            val_num = T_alg[j]
            val_ref = T_ana[j]
            
            if not np.isclose(val_num, val_ref, atol=1e-9):
                print(f"T_{j} FAIL            | {n:<5} | {val_num:.5e}    | {val_ref:.5e}    | ERRO")
                all_passed = False

    print("-" * 75)
    if all_passed:
        print("Todos os coeficientes V e T (até ordem 3) coincidem com a analítica.")
    else:
        print("Divergências encontradas.")


if __name__ == "__main__":
    run_VT_validation_tests()
```

## Parte 4: Integrais de Dotsenko-Fateev

Com os coeficientes e as funções $V$ e $T$ definidas, resta ainda definir as integrais de DF, além de suas primeira derivadas. De forma geral, as integrais são obtidas em termos de hipergeométricas como:

$$\begin{align}
	\mathcal{I}(\alpha,\beta,\gamma,\omega_n) = &-\sin(\pi \alpha)\tilde{I}_1(\alpha;\beta;\gamma;\omega_n)I_2(\alpha;\beta;\gamma;\omega_n)\nonumber\\
	&-\sin(\pi\beta)I_1(\alpha;\beta;\gamma;\omega_n)\tilde{I}_2(\alpha;\beta;\gamma;\omega_n).,
	\end{align}$$
onde

$$\begin{align}
		I_1(\alpha;\beta;\gamma;\omega_n) &= \frac{\Gamma(-1-\alpha-\beta-\gamma)\Gamma(1+\beta)}{\Gamma(-\alpha-\gamma)}F(-\gamma;-1-\alpha-\beta-\gamma;-\alpha-\gamma;\omega_n)\\
		I_2(\alpha;\beta;\gamma;\omega_n) &= \frac{\Gamma(1+\alpha)\Gamma(1+\gamma)}{\Gamma(2+\alpha+\gamma)}\omega_n^{1+\alpha+\gamma}F(-\beta;1+\alpha;2+\alpha+\gamma;\omega_n),\\
		\tilde{I}_1(\alpha;\beta;\gamma;\omega_n) &= I_1(\beta;\alpha;\gamma;1-\omega_n),\\
		\tilde{I}_2(\alpha;\beta;\gamma;\omega_n) &= I_2(\beta;\alpha;\gamma;1-\omega_n).
\end{align}$$

em que $\Gamma(x)$ é a função gamma, e $ F(a,b,c,z)$ a hipergeométrica de Gauss. Será necessário portanto obter as integrais de DF tanto para a integral "base" quanto para as integrais deslocadas. Para a integral base temos as funções $I_1,I_2,\tilde I_1$ e $\tilde I_2$:
$$
\begin{align}
	I_1(\alpha;\beta;\gamma;\omega_n) &= \frac{\pi k(n-k)}{2n^2}\omega_n^2\ {}_2F_1(2-\frac{k}{n};\frac{3}{2};3;\omega_n)\\
	I_2(\alpha;\beta;\gamma;\omega_n) &= \frac{\Gamma(2-\frac{k}{n})\Gamma(1+\frac{k}{n})}{2}\omega_n^2\ {}_2F_1(\frac{3}{2};2-\frac{k}{n};3;\omega_n)\\
	\tilde I_1(\alpha;\beta;\gamma;\omega_n) &= -\frac{2\sqrt{\pi}\Gamma(2-\frac{k}{n})}{\Gamma(\frac{3}{2}-\frac{k}{n})}\ {}_2F_1(-\frac{k}{n};-\frac{1}{2};\frac{3}{2}-\frac{k}{n};1-\omega_n)\\
	\tilde I_2(\alpha;\beta;\gamma;\omega_n) &= -\frac{2\sqrt{\pi}\Gamma(1+\frac{k}{n})}{\Gamma(\frac{1}{2}+\frac{k}{n})}(1-\omega_n)^{\frac{k}{n}-\frac{1}{2}}\ {}_2F_1(\frac{k}{n}-1;-\frac{1}{2};\frac{1}{2}+\frac{k}{n};1-\omega_n)
\end{align}
$$


Nota: É importante notar que, pela definição, a função $I_1$ possui uma divergencia devido a função gamma e a hipergeometrica, mas que pode ser realizada uma regularização que resulta na expressão acima. 

Nota 2: O mesmo argumento é utilizado ao notar que se $3/2-k/n$ for um inteiro negativo (ou 0), a função $\tilde I_1$ diverge. E portanto, neste caso, definindo $3/2-k/n=-m$, obtemos a expressão regularizada
$$\tilde I_1^{\text{reg}}(\alpha;\beta;\gamma;\omega_n) = -\frac{2\sqrt{\pi}\Gamma(2-\frac{k}{n})}{(m+1)!}\left(\frac{k}{n}\right)^{(m+1)}\left(\frac{1}{2}\right)^{(m+1)}(1-\omega_n)^{m+1}\ {}_2F_1(-\frac{1}{2};\frac{k}{n}-1;\frac{1}{2}+\frac{k}{n};1-\omega_n)$$

Para a integral deslocada no parametro $\alpha$ temos:

$$\begin{align}
	I_1(\alpha-s;\beta;\gamma;\omega_n) &= -\frac{2\sqrt{\pi}\Gamma(s-\frac{1}{2})}{\Gamma(s-1)}\ {}_2F_1(-\frac{k}{n};s-\frac{1}{2};s-1;\omega_n)\\
	I_2(\alpha-s;\beta;\gamma;\omega_n) &= \frac{\Gamma(1+\frac{k}{n})\Gamma(2-s-\frac{k}{n})}{\Gamma(3-s)}\omega_n^{2-s}\ {}_2F_1(\frac{3}{2};2-s-\frac{k}{n};3-s;\omega_n)\\
	\tilde I_1(\alpha-s;\beta;\gamma;\omega_n) &= \frac{\Gamma(s-\frac{1}{2})\Gamma(2-s-\frac{k}{n})}{\Gamma(\frac{3}{2}-\frac{k}{n})}\ {}_2F_1(-\frac{k}{n};s-\frac{1}{2};\frac{3}{2}-\frac{k}{n};1-\omega_n)\\
	\tilde I_2(\alpha-s;\beta;\gamma;\omega_n) &= -\frac{2\sqrt{\pi}\Gamma(1+\frac{k}{n})}{\Gamma(\frac{1}{2}+\frac{k}{n})}(1-\omega_n)^{\frac{k}{n}-\frac{1}{2}}\ {}_2F_1(s-1+\frac{k}{n};-\frac{1}{2};\frac{1}{2}+\frac{k}{n};1-\omega_n)
\end{align}$$

Aqui a análise de divergência é um pouco mais delicada. Quando $s=1$, a função $I_1$ diverge, e a regularização leva à:
$$I_1^{\text{reg}}(\alpha-1;\beta;\gamma;\omega_n) = \frac{{k\pi}}{n}\omega_n\ {}_2F_1(1-\frac{k}{n};\frac{3}{2};2;\omega_n)$$

Quando $s \geq 3$ a função $I_2$ diverge, e a função regularizada é dada por:
$$I_2^{\text{reg}}(\alpha-s;\beta;\gamma;\omega_n) = \frac{\Gamma(1+\frac{k}{n})\Gamma(2-s-\frac{k}{n})}{(s-2)!}\left(\frac{k}{n}+s-2\right)^{\underline{s-2}}\left(-\frac{3}{2}\right)^{\underline{s-2}}\ {}_2F_1(s-\frac{1}{2};-\frac{k}{n};s-1;\omega_n)$$

Por fim, para todos os valores de $s$, no caso em que $m=k/n-3/2$ é um inteiro, a função $\tilde I_1$ diverge e a regularização leva à:
$$\begin{align}
	\tilde I_1^{\text{reg}}(\alpha-s;\beta;\gamma;\omega_n) = &\frac{\Gamma(s-\frac{1}{2})\Gamma(2-s-\frac{k}{n})}{(m+1)!}\nonumber\\
	&\times\left(\frac{k}{n}\right)^{\underline{m+1}}\left(\frac{1}{2}-s\right)^{\underline{m+1}}(1-\omega_n)^{m+1}\ {}_2F_1(-\frac{1}{2};s+\frac{k}{n}-1;\frac{k}{n}+\frac{1}{2};1-\omega_n)
\end{align}$$

Finalmente, devemos fazer a mesma análise para as funções de DF deslocadas no parâmetro $\gamma$. Em que obtemos:

$$\begin{align}
	I_1(\alpha;\beta;\gamma-r;\omega_n) &= -\frac{2\sqrt{\pi}\Gamma(r-\frac{1}{2})}{\Gamma(r-1)}\ {}_2F_1(r-\frac{k}{n};r-\frac{1}{2};r-1;\omega_n)\\
	I_2(\alpha;\beta;\gamma-r;\omega_n) &= \frac{\Gamma(2-\frac{k}{n})\Gamma(1-r+\frac{k}{n})}{\Gamma(3-r)}\omega_n^{2-r}\ {}_2F_1(\frac{3}{2};2-\frac{k}{n};3-r;\omega_n)\\
	\tilde I_1(\alpha;\beta;\gamma-r;\omega_n) &= \frac{\Gamma(r-\frac{1}{2})\Gamma(2-\frac{k}{n})}{\Gamma(\frac{3}{2}+r-\frac{k}{n})}\ {}_2F_1(r-\frac{k}{n};r-\frac{1}{2};r+\frac{3}{2}-\frac{k}{n};1-\omega_n)\\
	\tilde I_2(\alpha;\beta;\gamma-r;\omega_n) &= -\frac{2\sqrt{\pi}\Gamma(1-r+\frac{k}{n})}{\Gamma(\frac{1}{2}-r+\frac{k}{n})}(1-\omega_n)^{\frac{k}{n}-r-\frac{1}{2}} {}_2F_1(\frac{k}{n}-1;-\frac{1}{2};\frac{1}{2}-r+\frac{k}{n};1-\omega_n)
\end{align}$$

Essas funções são ainda mais delicadas que o caso anterior, dado que agora o terceiro argumento da hipergeométrica tambem depende de $r$. Para o caso $r=1$ a função $I_1$ diverge e a regularização leva à
$$I_1^{\text{reg}}(\alpha;\beta;\gamma-1;\omega_n) = -\frac{{\pi}(n-k)}{n}\omega_n\ {}_2F_1(2-\frac{k}{n};\frac{3}{2};2;\omega_n)$$

Para a função $I_2$, a divergência ocorre sempre que $r\geq 3$, e a regularização leva à:
$$	I_2^{\text{reg}}(\alpha;\beta;\gamma-r;\omega_n) = \frac{\Gamma(2-\frac{k}{n})\Gamma(1+\frac{k}{n}-r)}{(r-2)!}\left(-\frac{3}{2}\right)^{\underline{r-2}}\left(\frac{k}{n}-2\right)^{\underline{r-2}}\ {}_2F_1(r-\frac{1}{2};r-\frac{k}{n};r-1;\omega_n)$$

A divergência da função $\tilde I_1$ ocorre sob duas condições: $m=k/n-3/2$ deve ser um inteiro e $m\geq r$. No caso das duas condições serem satisfeitas, a função deve então ser regularizada de forma que:
$$\begin{align}
\tilde I_1^{\text{reg}}(\alpha;\beta;\gamma-r;\omega_n) = &\frac{\Gamma(r-\frac{1}{2})\Gamma(2-\frac{k}{n})}{(m-r+1)!}\left(\frac{k}{n}-r \right)^{\underline{m-r+1}}\left( \frac{1}{2}-r\right)^{\underline{m-r+1}}\nonumber\\
&\times(1-\omega_n)^{m-r+1}\ {}_2F_1(-\frac{1}{2};\frac{k}{n}-1;m-r+{2};1-\omega_n)
\end{align}$$

Por fim, para $\tilde I_2$ o processo é similar, porém, a divergência ocorre se $\ell=k/n+1/2$ é um inteiro e $r\geq \ell$. E neste caso, a regularização leva à:
$$\begin{align} 
	\tilde I_2^{\text{reg}}(\alpha;\beta;\gamma-r;\omega_n) = &-\frac{2\sqrt{\pi}\Gamma(1-r+\frac{k}{n})}{(r-\ell+1)!} \left(1-\frac{k}{n} \right)^{\underline{r-\ell+1}}\left( \frac{1}{2}\right)^{\underline{r-\ell+1}}\nonumber\\
	 &\times {}_2F_1\left(r-\frac{1}{2}; r-\frac{k}{n}; r-\ell+2; 1-\omega_n\right) 
	 \end{align}$$

Devemos portanto, levar em consideração todos esses casos ao definirmos as funções $I$'s que serão utilizadas para obter a integral de correção da dimensão conforme. 


```python
## As seguintes funções implementam todas as considerações anteriores. Separando nas situações em que temos as integrais de DF 'base',
## as integrais deslocadas no parametro "alpha" e "gamma".

def is_close_int(val, tol=1e-9):
    """
    Essa função verifica se certo valor é um número inteiro com uma tolerância de 1e-9
    """
    r = round(val)
    return abs(val - r) < tol, int(r)

# Todas as funções abaixo definidas retornam a seguinte coleção de resultados:
# pre_factor: É o termo multiplicativo das funções I
# exp_w: em funções do tipo \omega_n ^{p}, esse valor indica o expoente do monômio.
# exp_1mw: em funções do tipo (1-\omega_n )^{p}, esse valor indica o expoente do monômio.
# A, B, C: Os valores indicam os argumentos da função hipergeométrica F(A,B,C,z) 
# arg_is_1_minus_w: O ultimo parâmetro que a função retorna é se a função I é do tipo F(A,B,C,omega_n) ou F(A,B,C,1-\omega_n).
# retornando False no primeiro caso e True no segundo.



def I1(n, k, case_type, val):
    """
    Implementa a função I_1 para os três casos na solução geral: "base", shift em "alpha" e shift em "gamma".
    val = valor de r ou s associado ao shift nos parametros.
    """
    
    kn = k / n
    
    
    # CASO BASE
    if case_type == 'base':
        pre = (np.pi * k * (n - k)) / (2 * n**2)
        return pre, 2, 0, 2 - kn, 1.5, 3, False


    # CASO ALPHA (Shift em alpha: val = s)
    elif case_type == 'alpha':
        s = val
        # Singularidade em s = 1
        if s == 1:
            # Regularizado
            pre = (k * np.pi) / n
            return pre, 1, 0, 1 - kn, 1.5, 2, False
        else:
            # Padrão para s != 1
            pre = -(2 * np.sqrt(np.pi) * sc.gamma(s - 0.5)) / sc.gamma(s - 1)
            return pre, 0, 0, -kn, s - 0.5, s - 1, False


    # CASO GAMMA (Shift em gamma: val = r)
    elif case_type == 'gamma':
        r = val
        # Singularidade em r = 1
        if r == 1:
            # Regularizado
            pre = -(np.pi * (n - k)) / n
            return pre, 1, 0, 2 - kn, 1.5, 2, False
        else:
            # Padrão para r != 1
            pre = -(2 * np.sqrt(np.pi) * sc.gamma(r - 0.5)) / sc.gamma(r - 1)
            return pre, 0, 0, r - kn, r - 0.5, r - 1, False
            
    return None

def I2(n, k, case_type, val):
    """
    Implementa a função I_2 para os três casos na solução geral: "base", shift em "alpha" e shift em "gamma".
    val = valor de r ou s associado ao shift nos parametros.
    """
    kn = k / n
    

    # CASO BASE
    if case_type == 'base':
        pre = (sc.gamma(2 - kn) * sc.gamma(1 + kn)) / 2.0
        return pre, 2, 0, 1.5, 2 - kn, 3, False

    # CASO ALPHA (Shift em alpha: val = s)
    elif case_type == 'alpha':
        s = int(val)
        # Divergência se s >= 3
        if s >= 3:
            # Regularizado
            ff_term = falling_factorial(kn + s - 2, s - 2)
            ff_beta = falling_factorial(-1.5, s - 2)
            fact_s2 = sc.factorial(s - 2)
            
            pre = (sc.gamma(1 + kn) * sc.gamma(2 - s - kn)) / fact_s2
            pre *= ff_term * ff_beta
            
            return pre, 0, 0, s - 0.5, -kn, s - 1, False
        else:
            # Padrão (s=1, 2)
            pre = (sc.gamma(1 + kn) * sc.gamma(2 - s - kn)) / sc.gamma(3 - s)
            return pre, 2 - s, 0, 1.5, 2 - s - kn, 3 - s, False


    # CASO GAMMA (Shift em gamma: val = r)
    elif case_type == 'gamma':
        r = int(val)
        # Divergência se r >= 3
        if r >= 3:
            # Regularizado
            ff_beta = falling_factorial(-1.5, r - 2)
            ff_term = falling_factorial(kn - 2, r - 2)
            fact_r2 = sc.factorial(r - 2)
            
            pre = (sc.gamma(2 - kn) * sc.gamma(1 + kn - r)) / fact_r2
            pre *= ff_beta * ff_term
            
            return pre, 0, 0, r - 0.5, r - kn, r - 1, False
        else:
            # Padrão (r=1, 2)
            pre = (sc.gamma(2 - kn) * sc.gamma(1 - r + kn)) / sc.gamma(3 - r)
            return pre, 2 - r, 0, 1.5, 2 - kn, 3 - r, False

    return None

def It1(n, k, case_type, val):
    """
    Implementa a função \tilde I_1 para os três casos na solução geral: "base", shift em "alpha" e shift em "gamma".
    val = valor de r ou s associado ao shift nos parametros.
    """
    kn = k / n
    
    # Verificação da condição de divergência: 3/2 - k/n = -m (m inteiro >= 0)
    # Ou seja: k/n - 3/2 = m
    m_check_val = kn - 1.5
    is_m_int, m = is_close_int(m_check_val)
    if not is_m_int or m < 0:
        is_m_int = False # Só diverge se m for inteiro não negativo
        m = 0

    # CASO BASE
    if case_type == 'base':
        if is_m_int:
            # Regularizado
            fact_m1 = sc.factorial(m + 1)
            ff_kn = falling_factorial(kn, m + 1)
            ff_half = falling_factorial(0.5, m + 1)
            
            pre = -(2 * np.sqrt(np.pi) * sc.gamma(2 - kn)) / fact_m1
            pre *= ff_kn * ff_half
            
            return pre, 0, m + 1, -0.5, kn - 1, 0.5 + kn, True
        else:
            # Padrão
            pre = -(2 * np.sqrt(np.pi) * sc.gamma(2 - kn)) / sc.gamma(1.5 - kn)
            return pre, 0, 0, -kn, -0.5, 1.5 - kn, True

    # CASO ALPHA (Shift em alpha: val = s)
    elif case_type == 'alpha':
        s = val
        if is_m_int:
            # Regularizado
            fact_m1 = sc.factorial(m + 1)
            ff_kn = falling_factorial(kn, m + 1)
            ff_half_s = falling_factorial(0.5 - s, m + 1)
            
            pre = (sc.gamma(s - 0.5) * sc.gamma(2 - s - kn)) / fact_m1
            pre *= ff_kn * ff_half_s
            
            return pre, 0, m + 1, -0.5, s + kn - 1, kn + 0.5, True
        else:
            # Padrão
            pre = (sc.gamma(s - 0.5) * sc.gamma(2 - s - kn)) / sc.gamma(1.5 - kn)
            return pre, 0, 0, -kn, s - 0.5, 1.5 - kn, True

    # CASO GAMMA (Shift em gamma: val = r)
    elif case_type == 'gamma':
        r = val
        # Diverge apenas se m é inteiro E m >= r
        diverges = is_m_int and (m >= r)
        
        if diverges:
            # Regularizado
            fact_mr1 = sc.factorial(m - r + 1)
            ff_kn_r = falling_factorial(kn - r, m - r + 1)
            ff_half_r = falling_factorial(0.5 - r, m - r + 1)
            
            pre = (sc.gamma(r - 0.5) * sc.gamma(2 - kn)) / fact_mr1
            pre *= ff_kn_r * ff_half_r
            
            return pre, 0, m - r + 1, -0.5, kn - 1, m - r + 2, True
        else:
            # Padrão
            pre = (sc.gamma(r - 0.5) * sc.gamma(2 - kn)) / sc.gamma(1.5 + r - kn)
            return pre, 0, 0, r - kn, r - 0.5, r + 1.5 - kn, True

    return None

def It2(n, k, case_type, val):
    """
    Implementa a função \tilde I_2 para os três casos na solução geral: "base", shift em "alpha" e shift em "gamma".
    val = valor de r ou s associado ao shift nos parametros.
    """
    kn = k / n
    
    # CASO BASE
    if case_type == 'base':
        # Sempre Convergente
        pre = -(2 * np.sqrt(np.pi) * sc.gamma(1 + kn)) / sc.gamma(0.5 + kn)
        return pre, 0, kn - 0.5, kn - 1, -0.5, 0.5 + kn, True

    # CASO ALPHA (Shift em alpha: val = s)
    elif case_type == 'alpha':
        s = val
        # Sempre Convergente
        pre = -(2 * np.sqrt(np.pi) * sc.gamma(1 + kn)) / sc.gamma(0.5 + kn)
        return pre, 0, kn - 0.5, s - 1 + kn, -0.5, 0.5 + kn, True

    
    # CASO GAMMA (Shift em gamma: val = r)
    elif case_type == 'gamma':
        r = val
        # Divergência se l = k/n + 1/2 for inteiro E r >= l
        l_check_val = kn + 0.5
        is_l_int, l_val = is_close_int(l_check_val)
        
        diverges = is_l_int and (r >= l_val)
        
        if diverges:
            # Regularizado
            fact_denom = sc.factorial(r - l_val + 1)
            ff_1mk = falling_factorial(1 - kn, r - l_val + 1)
            ff_half = falling_factorial(0.5, r - l_val + 1)
            
            pre = -(2 * np.sqrt(np.pi) * sc.gamma(1 - r + kn)) / fact_denom
            pre *= ff_1mk * ff_half
            
            return pre, 0, 0, r - 0.5, r - kn, r - l_val + 2, True
        else:
            # Padrão
            pre = -(2 * np.sqrt(np.pi) * sc.gamma(1 + kn - r)) / sc.gamma(0.5 + kn - r)
            return pre, 0, kn - r - 0.5, kn - 1, -0.5, 0.5 + kn - r, True

    return None
```

Por fim, precisamos ainda das primeiras derivadas das integrais de DF. Para isso utilizamos da derivada conhecida da hipergeometrica:

$$\frac{d}{dz} \ {}_2 F_1(a,b,c,z) = \frac{ab}{c} \ {}_2F_1(a+1,b+1,c+1,z) $$

Em geral, como discutido anteriormente, as funções $I$ serão dadas por expressões como:
$$ I = \text{pre} \ \times \omega_n^{\text{exp w}} \times (1-\omega_n)^{\text{exp 1mw}}\times \ {}_2 F_1(A,B,C,z)$$

sendo $z=\omega_n$ ou $1-\omega_n$. Assim, podemos definir uma função que recebe os parâmetros necessários e retorna a primeira derivada da função.

```python
def hyp2f1_val_der(a, b, c, z):
    """
    Dados os parametros A,B,C,Z a função retorna a hipergeométrica e a derivada da mesma.
    """
    val = sc.hyp2f1(a, b, c, z)
    der = (a * b / c) * sc.hyp2f1(a + 1, b + 1, c + 1, z)
    return val, der

def general_func_solver(w, pre_factor, exp_w, exp_1mw, A, B, C, arg_is_1_minus_w=False):
    if arg_is_1_minus_w:
        z = 1.0 - w
        dz_dw = -1.0
    else:
        z = w
        dz_dw = 1.0

    H, dH_dz = hyp2f1_val_der(A, B, C, z)
    term_w = w**exp_w
    term_1mw = (1.0 - w)**exp_1mw
    
    if exp_w == 0: d_term_w = 0.0
    else: d_term_w = exp_w * (w**(exp_w - 1))
    
    if exp_1mw == 0: d_term_1mw = 0.0
    else: d_term_1mw = -exp_1mw * ((1.0 - w)**(exp_1mw - 1))
    
    F_val = pre_factor * term_w * term_1mw * H

    ## A derivada final será obtida pela regra do produto entre os três termos funcções de \omega_n
    t1 = d_term_w * term_1mw * H
    t2 = term_w * d_term_1mw * H
    t3 = term_w * term_1mw * (dH_dz * dz_dw)
    F_der = pre_factor * (t1 + t2 + t3)
    
    return F_val, F_der
```

Com as funções $I$ e suas derivadas definidas, podemos assim obter as expressões para as integrais de DF e sua derivada. Pela definição dada anteriormente, a primeira derivada é obtida pela regra do produto por:

$$\begin{align}
	\frac{\partial}{\partial\omega_n}\mathcal{I}(\alpha;\beta;\gamma;\omega_n) = &-\sin(\pi\alpha)\left[I_2\partial_{\omega_n}\tilde{I}_1+\tilde{I}_1\partial_{\omega_n}I_2  \right]\nonumber\\
	&-\sin(\pi\beta)\left[\tilde{I}_2\partial_{\omega_n}{I}_1^{\text{reg}}+I_1^{\text{reg}}\partial_{\omega_n}\tilde{I}_2  \right]
\end{align} \tag{3}$$

E portanto, definimos a função que recebe os valores de $n,k$ e o shift em $\alpha$ ou $\gamma$ e retorna a integral de DF e sua derivada. 

```python
def Dotsenko_Fateev_Engine(n, k,case_type, val):
    w_n = omega(n)
    
    p1 = I1(n,k, case_type, val)
    p2 = I2(n,k, case_type, val)
    pt1 = It1(n,k, case_type, val)
    pt2 = It2(n, k,case_type, val)
    

    v1, d1 = general_func_solver(w_n, *p1)
    v2, d2 = general_func_solver(w_n, *p2)
    vt1, dt1 = general_func_solver(w_n, *pt1)
    vt2, dt2 = general_func_solver(w_n, *pt2)
    
    alpha = 1.0 - k/n
    beta = -1.5
    if case_type == 'alpha': alpha -= val
    
    sin_a = np.sin(np.pi * alpha)
    sin_b = np.sin(np.pi * beta)
    
    # Montagem Final da Integral
    val_A = vt1 * v2
    der_A = dt1 * v2 + vt1 * d2
    
    val_B = v1 * vt2
    der_B = d1 * vt2 + v1 * dt2
    
    I_final = -sin_a * val_A - sin_b * val_B
    I_deriv_final = -sin_a * der_A - sin_b * der_B
    
    return I_final, I_deriv_final
```

Novamente, é um exercicio de boas práticas verificar se nosso algoritmo calcula corretamente as derivadas das integrais de Dotsenko-Fateev. Para isso, precisamos computar analiticamente uma integral e sua derivada e construir um algoritmo de validação. Para $k=1$ do caso base, temos as derivadas das funções $I$'s:

$$\begin{align}
	\partial_{\omega_n}I_1^{\text{reg}} &= \frac{\pi(n-1)}{2n^2}\left[ 2\omega_n\ {}_2F_1(2-\frac{1}{n};\frac{3}{2};3;\omega_n) +\omega_n^2\frac{2n-1}{2n}\ {}_2F_1(3-\frac{1}{n};\frac{5}{2};4;\omega_n) \right]\\
	\partial_{\omega_n}I_2 &= \frac{\Gamma(2-\frac{1}{n})\Gamma(1+\frac{1}{n})}{2!}\left[2\omega_n\ {}_2F_1(\frac{3}{2};2-\frac{1}{n};3;\omega_n)+\omega_n^2\frac{2n-1}{2n}\ {}_2F_1(\frac{5}{2};3-\frac{1}{n};4;\omega_n)\right]\\
	\partial_{\omega_n}\tilde I_1 &= \frac{2\sqrt{\pi}\Gamma(2-\frac{1}{n})}{\Gamma(\frac{3}{2}-\frac{1}{n})}\frac{1}{3n-2}\ {}_2F_1(1-\frac{1}{n};\frac{1}{2};\frac{5}{2}-\frac{1}{n};1-\omega_n)\\
	\partial_{\omega_n}\tilde I_2 &= -\frac{2\sqrt{\pi}\Gamma(1+\frac{1}{n})}{\Gamma(\frac{1}{2}+\frac{1}{n})}(1-\omega_n)^{-\frac{1}{2}+\frac{1}{n}}\nonumber\\
	&\hspace{1cm}\times\Bigg[ \frac{n-2}{2n}(1-\omega_n)^{-1}\ {}_2F_1(-1+\frac{1}{n};-\frac{1}{2};\frac{1}{2}+\frac{1}{n};1-\omega_n)\nonumber\\
	&\hspace{2cm}-\frac{n-1}{n+2}\ {}_2F_1(\frac{1}{n};\frac{1}{2};\frac{3}{2}+\frac{1}{n};1-\omega_n)\Bigg]
\end{align}$$

Para isso utilizamos das regras de derivação de hipergeometricas e regra do produto. Por fim, podemos obter a derivada da integral de DF pela expressão (3) obtida anteriormente. A função abaixo implementa esse resultado analitico e compara, para uma coleção de valores de n, com o resultado gerado pelo algoritmo.

```python
def analytical_derivative_check(n):
    """
    Calcula analiticamente a derivada da integral de DF para k=1.
    """
    w = 4 * n / (n + 1)**2
    om = 1.0 - w 
    F = sc.hyp2f1
    G = sc.gamma

    # TERMO 1 e 2 da solução analitica
    sin_term = np.sin(np.pi/n)
    gamma_factor_1 = (G(2 - 1/n)**2 * G(1 + 1/n)) / G(1.5 - 1/n)
    C1 = np.sqrt(np.pi) * sin_term * gamma_factor_1
    
    term1 = (w**2 / (3*n - 2)) * F(1.5, 2 - 1/n, 3, w) * F(1 - 1/n, 0.5, 2.5 - 1/n, om)
    
    hyp_base_tilde = F(-1/n, -0.5, 1.5 - 1/n, om)
    brac1_part1 = 2 * w * F(1.5, 2 - 1/n, 3, w)
    brac1_part2 = (w**2) * ((2*n - 1)/(2*n)) * F(2.5, 3 - 1/n, 4, w)
    term2 = hyp_base_tilde * (brac1_part1 + brac1_part2)
    
    Block1 = C1 * (-term1 + term2) 

    # TERMO 3 e 4 da solução analitica
    gamma_factor_2 = (np.pi**1.5 * (n - 1) * G(1 + 1/n)) / (n**2 * G(0.5 + 1/n))
    C2 = gamma_factor_2
    pow_om = om**(1/n - 0.5)
    
    hyp_base_tilde_2 = F(-1 + 1/n, -0.5, 0.5 + 1/n, om)
    brac2_part1 = 2 * w * F(2 - 1/n, 1.5, 3, w)
    brac2_part2 = (w**2) * ((2*n - 1)/(2*n)) * F(3 - 1/n, 2.5, 4, w)
    term3 = pow_om * hyp_base_tilde_2 * (brac2_part1 + brac2_part2)
    
    hyp_base = F(2 - 1/n, 1.5, 3, w)
    big_brac_p1 = ((n - 2)/(2*n)) * (om**-1) * F(-1 + 1/n, -0.5, 0.5 + 1/n, om)
    big_brac_p2 = -((n - 1)/(n + 2)) * F(1/n, 0.5, 1.5 + 1/n, om)
    term4 = (w**2) * pow_om * hyp_base * (big_brac_p1 + big_brac_p2)
    
    Block2 = C2 * (term3 + term4)
    
    return Block1 + Block2

def run_validation():
    # Coleção de valores de n para verificar
    n_values = [2, 3, 4, 5, 10, 20, 50, 100]
    all_passed = True
    
    print(f"Iniciando verificação para n = {n_values} ...")
    print("-" * 60)

    for n in n_values:
        # Numérico (k=1)
        _, val_numeric_der = Dotsenko_Fateev_Engine(n, 1, 'base', 0)
        
        # Analítico
        val_analytic_der = analytical_derivative_check(n)
        
        # Comparação
        if not np.isclose(val_numeric_der, val_analytic_der, rtol=1e-9):
            all_passed = False
            print(f"\n FALHA para n={n}:")
            print(f"   Numérico:  {val_numeric_der:.8e}")
            print(f"   Analítico: {val_analytic_der:.8e}")
            print(f"   Diferença: {abs(val_numeric_der - val_analytic_der):.8e}")

    if all_passed:
        print("\n Os valores da derivada da integral de DF conferem com o calculado analítico para todos os valores testados.")
    else:
        print("\n Foram encontradas divergências.")

if __name__ == "__main__":
    run_validation()
```

## Parte 5: Integrais de Renormalização
Por fim, podemos finalmente reunir todos os algoritmos construidos anteriormente para obter a integral de renormalização. Para isso, utilizamos da expressão (2) e realizamos a soma sobre os blocos associados às diferentes integrais de DF.

```python
def kappa(n):
    """
    Este termo multiplica todos os blocos da integral final.
    """
    return ((n + 1) / (32 * n))**2

def J_3k_Final(n, k):
    """
    Calcula a Função de Correlação J_{3,k} completa para um dado n e k.
    """
    w_n = omega(n)
    alpha, gamma = alpha_gamma(n,k)
    val_kappa = kappa(n)
    

    # PARTE 1: TERMOS BASE (r=0, s=0)

    # Gerar V e T base
    gen_base = VT_Shifted_Generator(n,k, r=0, s=0)
    gen_base.generate_up_to(k)
    V_base, T_base = gen_base.get_numerical_values(w_n)
    
    # Integrais Base
    I_base, dI_base = Dotsenko_Fateev_Engine(n,k, 'base', 0)
    
    # Termo 1.1: [k^2 + 2kQ1/alpha]
    val_Q1 = Q_r(n, k, 1)
    term_bracket_base = (k**2) + (2 * k * val_Q1 / alpha)
    
    # Termo 1.2: Soma Omega_j * V_j
    sum_Omega_V = 0.0
    sum_Omega_T = 0.0
    
    for j in range(1, k + 1):
        O_j = Omega_m(n, k, j)
        sum_Omega_V += O_j * V_base[j]
        sum_Omega_T += O_j * T_base[j]
        
    # Montagem do Bloco Base
    # {Bracket + Sum(Omega*V)} * I  +  Sum(Omega*T) * dI
    Block_Base = (term_bracket_base + sum_Omega_V) * I_base + (sum_Omega_T * dI_base)


    # PARTE 2: GAMMA SHIFT (s varia, r=0)

    Block_Gamma = 0.0
    
    for s in range(1, k + 1):
        # Gera polinômios deslocados em s
        gen_s = VT_Shifted_Generator(n,k, r=0, s=s)
        max_j = k - s
        gen_s.generate_up_to(max_j) # Precisamos até k-s
        V_s, T_s = gen_s.get_numerical_values(w_n)
        
        # Integral deslocada em gamma
        I_gamma, dI_gamma = Dotsenko_Fateev_Engine(n,k, 'gamma', s)
        
        # Coeficiente Xi_s = Theta_{0,s}/2
        val_Xi = Theta_js(n, k, 0, s) / 2.0
        
        # Somas internas sobre j
        sum_Theta_V = 0.0
        sum_Theta_T = 0.0
        
        for j in range(1, k - s + 1):
            val_Theta = Theta_js(n, k, j, s)
            sum_Theta_V += val_Theta * V_s[j]
            sum_Theta_T += val_Theta * T_s[j]
            
        # {Xi + Sum(Theta*V)} * I + Sum(Theta*T) * dI
        term_s = (val_Xi + sum_Theta_V) * I_gamma + (sum_Theta_T * dI_gamma)
        Block_Gamma += term_s


    # PARTE 3: ALPHA SHIFT (r varia, s=0)
    # Termos com Lambda e L_{r,j,i}

    Block_Alpha = 0.0
    
    for r in range(1, k + 1):
        # Gera polinômios deslocados em r
        gen_r = VT_Shifted_Generator(n,k, r=r, s=0)
        max_j = k - r
        gen_r.generate_up_to(max_j) 
        V_r, T_r = gen_r.get_numerical_values(w_n)
        
        # Integral deslocada em alpha
        I_alpha, dI_alpha = Dotsenko_Fateev_Engine(n,k, 'alpha', r)
        
        val_Lambda = Lambda_r(n, k, r)
        
        # Somas duplas internas (j e i)
        sum_L_V = 0.0
        sum_L_T = 0.0
        
        for j in range(1, k - r + 1):
            for i in range(1, j + 1):
                val_L = L_rji(n, k, r, j, i)
                
                sum_L_V += val_L * V_r[i]
                sum_L_T += val_L * T_r[i]
        
        # {Lambda + Sum(L*V)} * I + Sum(L*T) * dI
        term_r = (val_Lambda + sum_L_V) * I_alpha + (sum_L_T * dI_alpha)
        Block_Alpha += term_r
        
    Total_Sum = Block_Base + Block_Gamma + Block_Alpha
    return val_kappa * Total_Sum
```

## Parte 6: Análises e Resultados

O gráfico anterior mostra uma certo comportamento geral para as integrais de renormalização para diversos valores de $k$. Além disso, as integrais parecem ter um limite assintótico em que $\mathcal{J}\rightarrow 0$ quando $n\rightarrow \infty$. Voltemos a definição da integral de correção da dimensão conforme supondo a ausência de modos fracionários de simetria-R:

$$\mathcal{J}_{\pm}(n)=\int d^2u G^{\{\text{prim}\}}_n(u,\bar u)$$

$$ G^{\{\text{prim}\}}_n(u)=\Big\langle  \sigma^{\pm\dagger}_{(n)}(\infty)   O^{(\text{int})\dagger}_{(2)}(1)  O^{(\text{int})}_{(2)}(u)\sigma^{\pm}_{(n)} (0)\Big\rangle$$

O operador quiral $\sigma^{\pm}$ é conhecidamente protegido, o que significa dizer que a dimensão conforme não é corrigida, e portanto, a integral acima é identicamente nula para qualquer valor de n. De acordo com a atuação do modo $J^{3}_0$ nos quiráis primários, temos:
$$J^3_0\sigma^{\pm}_n=q_n\sigma^{\pm}_n$$
em que $q_n=(n\pm 1)/2$. Ou seja, é esperado que, no limite em que $n>>k$, o modo $J^3_{-k/n}$ se comporte como um modo zero, fazendo com que a função de correlação envolvendo os fracionarios seja simplificada por:
$$\Big\langle J^3_0 \sigma^{\dagger\pm}_{(n)}(\infty)  O^{(\text{int})\dagger}_{(2)}(1)  O^{(\text{int})}_{(2)}(u)J^3_0 \sigma^{\pm}_{(n)} (0)\Big\rangle=-q_n^2\Big\langle  \sigma^{\pm\dagger}_{(n)}(\infty)   O^{(\text{int})\dagger}_{(2)}(1)  O^{(\text{int})}_{(2)}(u) \sigma^{\pm}_{(n)} (0)\Big\rangle$$

e portanto, a integral de renormalização se anula. Isso nos demonstra que o comportamento das integrais obtido no gráfico anterior está de acordo com o esperado. Porém, podemos ainda realizar uma análise que nos indique a lei de potência em que a integral de renormalização corresponde nesse limite $n>>k$.

Vamos supor a princípio que a integral seja descrita por uma função $f(n)$:
$$f(n) = an^{-b} + c$$
e realizar o fit para uma coleção de valores de $k$ e $n$.

```python
from scipy.optimize import curve_fit


def power_law_model(n, a, b, c):
    """
    Modelo: f(n) = a * n^(-b) + c
    """
    return a * np.power(n, -b) + c

def fit_multiple_k(k_values, n_min_global=5, n_max=50):
    """
    Calcula e ajusta a curva J(n) para múltiplos valores de k.
    Retorna os parâmetros e plota os resultados comparativos.
    """
    

    fig, ax = plt.subplots(figsize=(10, 6))
    colors = ['#1f77b4', '#d62728', '#2ca02c', '#9467bd', '#ff7f0e', '#8c564b']
    
    print(f"{'k':<3} | {'Função Ajustada':<50} | {'$R^2$'}")
    print("-" * 80)

    for idx, k in enumerate(k_values):
        color = colors[idx % len(colors)]
        
        # Definimos um intervalo de dados em que n>k
        
        start_n = max(n_min_global, k + 1)
            
        n_data = np.arange(start_n, n_max + 1)
        y_data = []
        
        # utilizamos da função construida para obter os pontos do eixo y
        for n in n_data:
            try:
                val = J_3k_Final(n, k)
                y_data.append(val)
            except:
                y_data.append(np.nan)
        
        y_data = np.array(y_data)
        
        mask = ~np.isnan(y_data)
        n_clean = n_data[mask]
        y_clean = y_data[mask]
        
        if len(n_clean) < 5:
            print(f"{k:<3} | Pontos insuficientes para fit.")
            continue

        # Ajuste da Curva 
        p0 = [y_clean[0], 1.5, 0.0]
        
        try:
            popt, pcov = curve_fit(power_law_model, n_clean, y_clean, p0=p0, maxfev=20000)
            a_fit, b_fit, c_fit = popt
            
            # Cálculo do R^2 para avaliar a qualidade
            residuals = y_clean - power_law_model(n_clean, *popt)
            ss_res = np.sum(residuals**2)
            ss_tot = np.sum((y_clean - np.mean(y_clean))**2)
            r_squared = 1 - (ss_res / ss_tot)
            
            func_str = f"f(n) = ({a_fit:.4f}) * n^(-{b_fit:.4f}) + ({c_fit:.4f})"
            
            print(f"{k:<3} | {func_str:<50} | {r_squared:.4f}")
            

            # Pontos discretos
            ax.scatter(n_clean, y_clean, color=color, s=25, alpha=0.6, label=f'Dados k={k}')
            
            # Linha do Fit
            n_smooth = np.linspace(n_clean.min(), n_clean.max(), 100)
            y_smooth = power_law_model(n_smooth, *popt)
            ax.plot(n_smooth, y_smooth, color=color, linestyle='--', linewidth=2, 
                    label=f'Fit k={k} ($b={b_fit:.2f}$)')
            
        except Exception as e:
            print(f"{k:<3} | Falha no Fit: {str(e)}")


    ax.set_xlabel('$n$')
    ax.set_ylabel(r'$\mathcal{J}_{3,k}$')
    ax.set_title('Ajuste de Lei de Potência: $f(n) = A n^{-b} + C$')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


k_lista = [1, 2, 3, 4]


fit_multiple_k(k_lista, n_min_global=5, n_max=50)
```

Com este resultado podemos notar uma certa constancia para o expoente que descreve a integral para valores grandes de n, sendo a função descrita, aproximadamente, por:
$$\mathcal{J}_{3,k}(n) \propto n^{-2}$$

Veja que no grafico, escolhemos um range de $n$ sempre maior do que $k$. Mas para valores como $k=4$ ou $k=5$ o range de n está proximo de $k$. Essa proximidade se reflete no coeficiente de ajuste $R^2$, que diminui quando $k$ se aproxima de $n$.

É nítido que para um valor fixo de $n$, a integral de renormalização cresce negativamente em função de $k$. Voltando a definição da correção da dimensão conforme $\Delta$ do campo descendente, nós obtivemos que, em segunda ordem de uma abordagem perturbativa sob um parâmetro $\lambda$, a dimensão conforme de um dado operador é renormalizada devido a inserção do operador de deformação como:

$$\Delta_{\lambda}=\Delta -\lambda^2 \frac{\pi}{2}\mathcal{J} + \mathcal{O}(\lambda^3) \approx \Delta\left( 1 +\lambda^2 \frac{\pi}{2}\left|\frac{\mathcal{J}}{\Delta}\right| \right)$$

É notável que, para um $k$ fixo, a correção da dimensão conforme ($\mathcal{J}$) diminui com n, e portanto, a correção perturbativa permanece válida. Porém, no caso de $n$ fixo, devemos verificar se a razão ${\mathcal{J}}/{\Delta}$ permanece finita ou se também cresce indefinidamente. No segundo caso, é possível que a abordagem perturbativa falhe, e portanto, devemos definir uma relação entre a escolha de um parametro $\lambda$ e um possível $k_{max}$ em que ainda seja permitido realizar o processo perturbativo.

Para o campo descendente $J^{3}_{-k/n}\sigma^{-}_n$ temos a dimensão conforme $\Delta$ dada por:
$$\Delta = n-1+\frac{2k}{n}$$

Podemos então definir um parâmetro $\epsilon$ que valida a teoria perturbativa indicando que a integral da renormalização é de fato uma correção da dimensão conforme, ou seja,
$$\lambda^2 \frac{\pi}{2}\left|\frac{\mathcal{J}}{\Delta}\right|< \epsilon$$

Assim, podemos fixar um valor de $n$, e verificar a razão entre a integral de correção e a dimensão conforme, de forma a obter valores de $k_{max}$ que se enquadram no limite perturbativo, baseado em uma escolha de acoplamento $\lambda$. 

```python

def check_perturbation_breakdown(n_fixed, k_max_scan=20):
    """
    Para um dado n, e um parâmetro corretivo epsilon, verifica para qual valor maximo de k a teoria perturbativa permanece valida. 
    """
    
    k_values = np.arange(1, k_max_scan + 1)
    ratios = []
    
    # Calculamos a Razão |J| / Delta
    for k in k_values:
        try:
            
            val_J = J_3k_Final(n_fixed, k)
            
            # Dimensão Conforme do descendente: Delta = n - 1 + 2k/n
            val_Delta = n_fixed - 1 + (2 * k / n_fixed)
            
            ratio = abs(val_J) / val_Delta
            ratios.append(ratio)
        except:
            ratios.append(np.nan)
            
    ratios = np.array(ratios)
    
    # Aqui definimos cenários de acoplamento (Lambda). 
    # Vamos testar lambda = 0.01, 0.1, 0.2, 0.5 
    lambdas = [0.01, 0.1, 0.2, 0.5]
    epsilon = 0.1 # Uma tolerância de 10% de correção
    
    plt.figure(figsize=(10, 6))
    
    # Plot da Razão 
   
    plt.plot(k_values, ratios, 'ko-', label=r' $|\mathcal{J}|/\Delta$')
    
    print(f"{'Lambda':<10} | {'k_max (Limite)'}")
    print("-" * 30)

    # Encontramos as interseções
    colors = ['green', 'blue', 'orange', 'red']
    for i, lam in enumerate(lambdas):
        # A correção total é: lambda^2 * (pi/2) * ratio
        correction = (lam**2) * (np.pi / 2) * ratios
        

        # Queremos que a correção seja menor que epsilon
        # Ou seja: ratio < 0.1 / (lambda^2 * pi/2)
        
        limit_val = epsilon / (lam**2 * np.pi / 2)
        
        # Desenhar linha limite no gráfico
        plt.axhline(y=limit_val, color=colors[i], linestyle='--', 
                    label=f'Limite para $\lambda={lam}$')
        
        # E marcamos o primeiro k que excede o limite
        # np.where retorna os indices onde a condição é falsa 
        broken_indices = np.where(ratios > limit_val)[0]
        
        if len(broken_indices) > 0:
            k_break = k_values[broken_indices[0]]
            print(f"{lam:<10} | {k_break}")
            # Marca no gráfico
            plt.plot(k_break, ratios[broken_indices[0]], 'X', color=colors[i], markersize=12)
        else:
            print(f"{lam:<10} | > {k_max_scan}")

    plt.yscale('log') 
    plt.title(f'Limite de Validade da Perturbação (n={n_fixed})')
    plt.xlabel('Modo k')
    plt.ylabel(r' $|\mathcal{J}| / \Delta$')
    plt.legend()
    plt.grid(True, which="both", alpha=0.3)
    plt.show()


check_perturbation_breakdown(n_fixed=20, k_max_scan=18)
```

Esse último resultado é algo ainda a ser discutido. A possível conclusão aqui é que para um dado conjunto de parametros $\epsilon$, $\lambda$ e $n$, há um valor máximo de $k$ para o descendente fracionário em que a integral $\mathcal{J}$ representa uma correção perturbativa.

Esse resultado poderia ser interpretado da seguinte forma: Os modos fracionários aumentam a dimensão conforme do campo quiral, e por consequência, temos um estado com maior energia. Em limites de altas energias, a abordagem perturbativa encontra uma limitação em termos do modo $k$.

Independentemente da validade da conclusão anterior, o algoritmo construido acima possui uma grande relevãncia para o contexto de lifting de descendentes fracionários. O mesmo procedimento para se obter a integral de forma analítica está sendo realizado para o caso de modos fracionários $J^{\pm}_{-k/n}$. Esses que possuem na literatura certos resultados tabelados para combinações específicas de $k$ e $n$. Sendo assim, a implemetnação de um algoritmo semelhante, pode ser útil para a discussão do tema.
